```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configuração inicial
BASE_DIR = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = BASE_DIR / "data" / "processed"

# Carrega datasets
df_brasileirao = pd.read_csv(DATA_DIR / "brasileirao_serie_a_clean.csv")
df_copa = pd.read_csv(DATA_DIR / "copa_brasil_clean.csv")

# 1. Estatísticas gerais
print("=== BRASILEIRÃO ===")
display(df_brasileirao.describe(include='all'))
print("\n=== COPA DO BRASIL ===")
display(df_copa.describe(include='all'))

# 2. Contagem de partidas por ano
plt.figure(figsize=(10,5))
df_brasileirao['ano_campeonato'].value_counts().sort_index().plot(kind='bar', color='royalblue')
plt.title("Número de Partidas por Ano - Brasileirão Série A")
plt.xlabel("Ano")
plt.ylabel("Partidas")
plt.show()

# 3. Médias de gols
media_gols = df_brasileirao.groupby("ano_campeonato")[["gols_mandante","gols_visitante"]].mean()
media_gols.plot(kind='line', figsize=(10,5))
plt.title("Média de Gols por Ano - Brasileirão Série A")
plt.ylabel("Gols Médios")
plt.show()

# 4. Taxa de vitórias por time (exemplo)
vit_mandante = df_brasileirao.groupby("time_mandante")["mandante_venceu"].mean().sort_values(ascending=False)
vit_mandante.head(10).plot(kind='bar', color='green', figsize=(10,5))
plt.title("Top 10 times com mais vitórias em casa")
plt.show()
```
